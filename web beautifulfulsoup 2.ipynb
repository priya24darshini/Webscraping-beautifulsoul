# 7) Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content
and the code for the video from the link for the youtube video from the post.

```python
from bs4 import BeautifulSoup
import requests

page=requests.get("https://coreyms.com/")
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
s_title=soup.find('a',class_="entry-title-link")
s_title
s_title.text
```

```python
titles=[]
for i in soup .find_all('a',class_="entry-title-link"):
    titles.append(i.text)
titles
```

```python
s_date=soup.find('time',class_="entry-time")
s_date
s_date.text
```

```python
dates=[]
for i in soup .find_all('time',class_="entry-time"):
    dates.append(i.text)
dates
```

```python
s_content=soup.find('div',class_="entry-content")
s_content
s_content.text
```

```python
contents=[]
for i in soup .find_all('div',class_="entry-content"):
    contents.append(i.text)
contents
```

```python
fr_cusine=soup.find('a',class_="/chennai-restaurants/south-indian-cuisine")
fr_cusine
fr_cusine.text

cusine=[]
for i in soup.find_all("a",class_="/chennai-restaurants/south-indian-cuisine"):
    cusine.append(i.text)
cusine
```

```python
 location=[]
for i in soup .find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
location
```
